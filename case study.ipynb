```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
data = pd.read_csv('manufacturing.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.jointplot(x='Number of Units',y='Manufacturing Cost',data=data)
```

```python
sns.lmplot(x='Number of Units',y='Manufacturing Cost',data=data)
```

```python
X = data[['Number of Units']]
y = data[['Manufacturing Cost']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression(fit_intercept=True)
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
sns.scatterplot(data=data, x = 'Number of Units',y='Manufacturing Cost')
plt.plot(X,reg.predict(X),c='r')
```

# Polynomial Model

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4)
```

```python
X_poly = poly.fit_transform(X)
```

```python
reg_poly = LinearRegression()
```

```python
reg_poly.fit(X_poly,y)
```

```python
reg_poly.coef_
```

```python
reg_poly.intercept_
```

```python
sns.scatterplot(data=data, x = 'Number of Units',y='Manufacturing Cost')
plt.plot(X,reg_poly.predict(X_poly),c='r')
```

```python
reg.score(X,y)
```

```python
reg_poly.score(X_poly,y)
```
